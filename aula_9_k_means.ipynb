```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class']
df = pd.read_csv('dataset/iris.data', header=None, names=columns)
df.head()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Petal length', y='Petal width', data=df, hue='Class', s=50)
```

```python
X = df.drop(columns = ['Sepal length', 'Sepal width','Class']).to_numpy()
y = df['Class']
```

```python
clf = KMeans()
clf.fit(X)
#Apenas para saber como K-Means agrupou
y_pred = clf.fit_predict(X)
centers = clf.cluster_centers_

plt.figure(figsize=(12,8))
plt.scatter(X[:,0], X[:,1], c=y_pred, s=50)
plt.scatter(centers[:,0], centers[:,1], c='red', s=50)
```

```python
clf = KMeans(n_clusters=3)
clf.fit(X)
#Apenas para saber como K-Means agrupou
y_pred = clf.fit_predict(X)
centers = clf.cluster_centers_

plt.figure(figsize=(12,8))
plt.scatter(X[:,0], X[:,1], c=y_pred, s=50)
plt.scatter(centers[:,0], centers[:,1], c='red', s=50)
```

```python
plt.figure(figsize=(18,18))
K = range(1,10)
for k in K:
    clf = KMeans(n_clusters=k, init='random', n_init=1)
    clf.fit(X)
    y_pred = clf.fit_predict(X)
    centers = clf.cluster_centers_

    plt.subplot(3,3,k)
    plt.scatter(X[:,0], X[:,1], c=y_pred, s=50)
    plt.scatter(centers[:,0], centers[:,1], c='red', s=50)

plt.show(True)
```

```python
WSS = [] #Método do Cotovelo para encontrar o melhor k
K = range(1,10)

for k in K:
    clf = KMeans(n_clusters=k)
    clf.fit(X)
    WSS.append(clf.inertia_)
    
plt.figure(figsize=(12,8))
plt.plot(K, WSS, '-bo')
```
