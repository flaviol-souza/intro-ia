```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.data', header=None)
df.rename(columns = {0:'Sepal length', 1:'Sepal width', 2:'Petal length', 3:'Petal width', 4:'Class'}, inplace = True)
df.head()
```

```python
df.plot.kde()
```

```python
df.plot(subplots=True, figsize=(10, 10));
```

```python
df['Class'].unique()
```

```python
colors = {'Iris-setosa':'red', 'Iris-versicolor':'green', 'Iris-virginica':'blue'}
```

```python
df.plot(x='Sepal length', y='Sepal width', kind='scatter',
        figsize=(10,6), c=df['Class'].map(colors),
        title='Sepal featuares in Class')
```

```python
df.plot(x='Petal length', y='Petal width', kind='scatter',
        figsize=(10,6), c=df['Class'].map(colors),
        title='Sepal featuares in Class')
```

```python
 df.plot.scatter(x='Petal length', y='Petal width',  figsize=(10,6),
                c=df['Class'].map(colors),
                s=df['Sepal length'] * 20)
```

```python
df.boxplot(column='Petal width', by='Class', figsize=(10,6))
```

```python
df.boxplot(by="Class", figsize=(10, 10))
plt.show()
```

```python
g = sns.pairplot(df, hue="Class")
```

```python
sns.jointplot(x="Petal length", y="Petal width", data=df, hue='Class')
```

```python
ax= sns.boxplot(x="Class", y="Petal length", data=df)
ax= sns.stripplot(x="Class", y="Petal length", data=df, jitter=True, edgecolor="gray")

#boxtwo = ax.artists[1]
#boxtwo.set_facecolor('red')
#boxtwo.set_edgecolor('black')
#boxthree=ax.artists[1]
#boxthree.set_facecolor('yellow')
#boxthree.set_edgecolor('black')

plt.show()
```

```python
iris_avg=df["Petal width"].groupby(df["Class"]).mean()
iris_avg.plot.pie(subplots=True,figsize=(10,10), autopct='%.2f', fontsize=20)
```
